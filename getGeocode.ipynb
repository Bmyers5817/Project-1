```python
import pandas as pd
from geopy.geocoders import Nominatim 
```

```python
df = pd.read_csv('data/2019.csv')[['Country or region','Score']]
df['Lat'] = ''
df['Lng'] = ''
df.head()
```

```python
geo = Nominatim(user_agent='World_Happiness')
```

```python
geo.geocode('Finland').latitude
```

```python
for index, row in df.iterrows():
    address = row['Country or region']
    try:
        print(f'Processing record {index} | {address}')
        df.loc[index, 'Lat'] = geo.geocode(address).latitude
        df.loc[index, 'Lng'] = geo.geocode(address).longitude
    except:
        print(f'{address} not found. Skipping....')
        pass
df
```

```python
df.to_csv('data/heatmap_data.csv')
```
