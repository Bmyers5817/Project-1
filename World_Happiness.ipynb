```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import seaborn as sn
```

```python
# Import and read .csv file
happiness_data_2015 = pd.read_csv("data/2015.csv")
happiness_data_2016 = pd.read_csv("data/2016.csv")
happiness_data_2017 = pd.read_csv("data/2017.csv")
happiness_data_2018 = pd.read_csv("data/2018.csv")
happiness_data_2019 = pd.read_csv("data/2019.csv")


happiness_data_2015.head()
```

```python
happiness_data_2015.insert(0, "Year", "2015", True) 

happiness_data_2015.drop(['Region', 'Standard Error', 'Dystopia Residual'], axis=1, inplace=True)
happiness_data_2015.rename({'Happiness Rank':'Overall rank', 
                           'Country':'Country or region', 
                           'Happiness Score': 'Score', 
                           'Economy (GDP per Capita)': 'GDP per capita', 
                           'Family':'Social support', 
                           'Health (Life Expectancy)':'Healthy life expectancy', 
                           'Freedom':'Freedom to make life choices', 
                           'Trust (Government Corruption)':'Perceptions of corruption'}, axis=1, inplace=True)





happiness_data_2015.head()
```

```python
happiness_data_2016.insert(0, "Year", "2016", True) 

happiness_data_2016.drop(['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'], axis=1, inplace=True)
happiness_data_2016.rename({'Happiness Rank':'Overall rank', 
                           'Country':'Country or region', 
                           'Happiness Score': 'Score', 
                           'Economy (GDP per Capita)': 'GDP per capita', 
                           'Family':'Social support', 
                           'Health (Life Expectancy)':'Healthy life expectancy', 
                           'Freedom':'Freedom to make life choices', 
                           'Trust (Government Corruption)':'Perceptions of corruption'}, axis=1, inplace=True)
happiness_data_2016.head()
```

```python
happiness_data_2017.insert(0, "Year", "2017", True)

happiness_data_2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia.Residual'], axis=1, inplace=True)
happiness_data_2017.rename({'Happiness.Rank':'Overall rank', 
                           'Country':'Country or region', 
                           'Happiness.Score': 'Score', 
                           'Economy..GDP.per.Capita.': 'GDP per capita', 
                           'Family':'Social support', 
                           'Health..Life.Expectancy.':'Healthy life expectancy', 
                           'Freedom':'Freedom to make life choices', 
                           'Trust..Government.Corruption.':'Perceptions of corruption'}, axis=1, inplace=True)
happiness_data_2017.head()
```

```python
happiness_data_2018.insert(0, "Year", "2018", True)
happiness_data_2019.insert(0, "Year", "2019", True)
```

```python
happiness_data = pd.concat([happiness_data_2019, happiness_data_2018, happiness_data_2017, happiness_data_2016, happiness_data_2015]).reset_index()

happiness_data.head(100)
```

```python
happiness_data["Country or region"].replace("Taiwan Province of China", "Taiwan", inplace = True)
happiness_data["Country or region"].replace("Hong Kong S.A.R., China", "Hong Kong", inplace = True)
happiness_data["Country or region"].replace("Somaliland region", "Somaliland Region", inplace = True)
happiness_data["Country or region"].replace("North Macedonia", "Macedonia", inplace = True)
happiness_data["Country or region"].replace("Northern Cyprus", "North Cyprus", inplace = True)
happiness_data["Country or region"].replace("Trinidad and Tobago", "Trinidad & Tobago", inplace = True)
del happiness_data['index']
happiness_data
```

```python
group_country = happiness_data.groupby("Country or region")
pd.set_option("display.max_rows", None, "display.max_columns", None)
group_country.count()
```

```python
country_count = happiness_data["Country or region"].value_counts()

happiness_data_drop = happiness_data[~happiness_data['Country or region'].isin(country_count[country_count < 5].index)]

happiness_data_final = happiness_data_drop.reset_index()
del happiness_data_final['index']
happiness_data_final
```

```python
# Group data by Country
gr_country = happiness_data_final.groupby(['Country or region'])
country = happiness_data_final['Country or region'].unique()
# Caluculate user defined statistics
mean_score = gr_country['Score'].mean()
mean_gdp = gr_country['GDP per capita'].mean()
mean_social = gr_country['Social support'].mean()
mean_health = gr_country['Healthy life expectancy'].mean()
mean_freedom = gr_country['Freedom to make life choices'].mean()
mean_generosity = gr_country['Generosity'].mean()
mean_corrupt = gr_country['Perceptions of corruption'].mean()
# Create dataframe with user defined statistics
summary_stats = pd.DataFrame({
   "Mean Score": mean_score,
   "Mean GDP": mean_gdp,
   'Mean Social Support': mean_social,
   "Mean Healthy Life Expectancy": mean_health,
   "Mean Freedom": mean_freedom,
   "Mean Generosity": mean_generosity,
   "Mean Corruption": mean_corrupt,
   })
summary_stats
```

```python
df = happiness_data_final
ranks = df.set_index('Overall rank')
ranks.head()

is_2015 = df['Year']=="2015"
is_2016 = df['Year']=="2016"
is_2017 = df['Year']=="2017"
is_2018 = df['Year']=="2018"
df_2015 = df[is_2015]
df_2016 = df[is_2016]
df_2017 = df[is_2017]
df_2018 = df[is_2018]
top_2015 = df_2015.iloc[0:10, 0:10]
top_2016 = df_2016.iloc[0:10, 0:10]
top_2017 = df_2017.iloc[0:10, 0:10]
top_2018 = df_2018.iloc[0:10, 0:10]
top_2019 = df.iloc[0:10, 0:10]
```

```python
labels = ["GDP", "Social", "Health", "Freedom", "Generosity", "Corruption"]

ave_gdp = happiness_data_final['GDP per capita'].mean()
ave_social = happiness_data_final['Social support'].mean()
ave_health = happiness_data_final['Healthy life expectancy'].mean()
ave_freedom = happiness_data_final['Freedom to make life choices'].mean()
ave_generosity = happiness_data_final['Generosity'].mean()
ave_corrupt = happiness_data_final['Perceptions of corruption'].mean()

sizes = [ave_gdp, ave_social, ave_health, ave_freedom, ave_generosity, ave_corrupt]

colors = ["crimson", "coral", "gold", "limegreen", "royalblue", "mediumpurple"]

explode = (0, 0.1, 0, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=90)
```

```python
# Scatterplot to see correlation between GDP and Life expectancy
plt.scatter(happiness_data_final["GDP per capita"], happiness_data_final["Healthy life expectancy"], marker="o", s=8)

plt.title("What is the Correlation between GDP per Capita and Life Expectancy?")
plt.xlabel("GDP per capita")
plt.ylabel("Healthy life expectancy")
plt.grid(axis='both')

plt.show()

corr_1 = round(st.pearsonr(happiness_data_final["GDP per capita"],happiness_data_final["Healthy life expectancy"])[0],2)
print(f"The correlation coefficient between GDP Per Captia and Healthy Life Expectancy is {corr_1}")
```
